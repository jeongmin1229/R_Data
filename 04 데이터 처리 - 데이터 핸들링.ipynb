### apply 계열의 함수들

```r
#apply() : 행단위 또는 열단위로 특정 함수를 적용할 떄 사용
#apply(행렬, 방향, 함수) : 반향 - 1(행), 2(열)

m = matrix(1:9, nrow=3)
m
apply(m, 1 ,sum) #행끼리의 합의 값
apply(m, 2 ,sum) #열끼리의 합의 값
```

```r
# iris의 각 관측값에 합을 계산하는 작업을 수행하시오
apply(iris[,-5], 2, sum)
colSums(iris[,-5]) #위와 동일
#행, 열의 평균을 계산하는 함수 : rowMeans(),colMeans()
#행, 열의 합을 계산하는 함수 : rowMeans(),colMeans()
```

```r
#lapply(벡터나 리스트 , 함수)

#만약에 관측값의 평균을 계산한 후 각 요소에 평균을 뺀값을 추출하시오.
b = c(10,13,21,33,40)
y=mean(b)
b-y

# lapply 함수로 다음과 같이 할 수 있다. (위와 동일)
f1= function(x){mean(b) - x}
result = lapply(b , f1 )
result

#unlist() : 리스트를 벡터로 변환
result1 = unlist(result)
result1
```

```r
x = list(a = 1:3, b = 4:6)
x

lapply(x, mean) # 리스트 또한 사용 가능

lapply(iris[,1:4],mean) #데이터 프레임도 가능
class(lapply(iris[,1:4],mean) ) #리스트로 반환된 결과가 나온다
colMeans(iris[,-5])
```

```r
# 데이터프레임을 lapply 함수를 이용해 계산된 결과는 리스트로 반환된다
# 리스트로 반환된 결과값을 데이터프레임으로 변환하려면 
# unlist() : 리스트를 벡터로 변환하는 함수
# matrix() : 벡터를 행렬로 변환하는 함수
# as.data.frame() : 행렬을 데이터프레임으로 형변환
# names() : 리스트의 결과에 변수를 데이터프레임의 열이름으로 부여하는 작업 

df = as.data.frame(matrix(unlist(lapply(iris[,1:4],mean)), byrow = T, ncol=4))
names(df) = names(iris[,-5])
df
```

```r
#do.call(호출할 함수, 파라미터)
#unlist()는 리스트를 벡터로 변환하는 함수이다. 벡터는 한가지 타입의 데이터만 저장할 수 있다.
#이에 반해 리스트는 서로다른 자료형을 담아서 처리할 수 있는 자료형이다.
#그래서 데이터가 손실이 될 수 도 있다.

data.frame(do.call(cbind, lapply(iris[,-5],mean)))
```

```r
#sapply : lapply와 유사하지만 리스트 대신 , 행렬 or 벡터 등으로 결과가 반환된다.
# iris 컴럼별로 평균을 출력

xy = sapply(iris[,-5],mean)
as.data.frame(t(xy)) # t() :전치행렬 = 행과 열을 바꿔줌

```

```r
#iris 데이터셋의 각 컬럼의 관측값이 3보다 큰값이인지 분석하시오.
k=sapply(iris[,-5], function(x){x>3})
k
class(k)
```

```r
#tapply(데이터, 색인, 함수) : 그룹별 처리를 위한 apply 계열의 함수 / 색인 : factor타입
tapply(1:10, rep(1:2,5), sum)

#1부터 10까지의 숫자의 짝수합과 짝수합을 산출
tapply(1:10, rep(c('홀','짝'),5), sum)

#A와 B그룹에 나이가 15~35세인 사람들이 5명씩 있을 때 그릅별 평균 연령을 산출하시오.
group = c(rep('A', 5), rep('B', 5))
age = c(sample(15:35, 10))
data = data.frame(group, age)
data
tapply(data$age, data$group, mean)
```

```r
#iris 데이터셋의 품종별로 Sepal.Length의 평균을 산출

tapply(iris$Sepal.Length, iris$Species, mean)
```
