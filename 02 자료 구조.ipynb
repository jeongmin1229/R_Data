## 4. 벡터 연산

```r
# 수정 
a = c(1,2,3,4,5)
a[3] = 10 # 3번째를 10으로 변경 
a[c(2,4)]= 9 # 2, 4번째를 9로 변경 
a
```

```r
# 5보다 큰 벡터의 요소의 값을 모두 3으로 변경 
a[a>5]=3
a
```

```r
# 2~5번째 요소의 값을 모두 0으로 변경 
a[2:5] = 0
a
```

```r
# Quiz
v_a = c(2:15)
v_a

v_a[] = 1
v_a
v_a[1:length(v_a)] = 1
v_a
```

```r
# 추가, c()
v_add = c(1,2,3,4,5)
```

```r
a = c(1,2,3)
b = c(4,5,6)
c = c(7,8,9)

new_v = c(a,b,c)
new_v
```

```r
# append(a,b): a,b 두개의 벡터를 연결하는 함수 
# append(a,b,index): index로 b벡터의 추가 위치 지정 가능 
v_a= c('A','B','C','F','G')
v_b= c('D','E')

append(v_a,v_b,3) # v_a의 3번째 위치에 v_b 벡터룰 추가함 
v_a
v_b
```

```r
# t_v의 마지막 요소를 제외하고 추출 
t_v[-length(t_v)]
```

```r
# !(NOT 연산자), 논리 부정 
logical_v = c(FALSE,TRUE,TRUE,FALSE,FALSE)
logical_v
!logical_v # 반대되는 값을 얻을 수 있음 
```

```r
# Fancy index: 팬시 인덱스, 논리조건식을 이용한 인덱스 처리 
v_str = c('첫번째','두번째','세번째','네번째','다섯번째')
v_str

# 두번째, 세번째 요소만 추출 
v_str[c(2,3)]
v_str[2:3]
```

```r
v_str[logical_v] # TRUE 위치에 있는 요소만 추출 
```

```r
# 벡터의 사칙연산: 벡터화 연산 (요소의 개수가 배수의 형태를 띄고 있어야함) 
a = c(1,2,3,4)
b = c(5,6,7,8)

# 더하기 (각각의 요소끼리 더해짐)
c = a+b
c

# 빼기
d = a-b
d

# 곱하기 
e = a * b
e

# 나누기 
f = a / b
f 
```

```r
a = c(1,2,3,4)
b = a + 2 # 각각의 요소에 2를 더함 
b
```

```r
a = c(1,2,3,4)
b = c(4,5)

# 경고가 발생하지만 계산은 수행됨 
a + b
```

```r
# %in% : ~안에 포함여부를 판정하는 연산자 
'a' %in% c('a','b','c')
'd' %in% c('a','b','c')
```

```r
# 벡터를 집합으로 취급, 집합연산자(합집합, 교집합, 차집합)
# 집합(set)은 중복을 허용하지 않는다 


# 합집합: union 
union(c('a','b','c'),c('a','d'))

# 교집합: intersect
intersect(c('a','b','c'),c('a','d'))

# 차집합: setdiff
setdiff(c('a','b','c'),c('a','d'))
```

```r
# 집합의 비교연산 equals
setequal(c('a','b','c'),c('a','d'))
setequal(c('a','b','c'),c('a','b','c','c'))
```

```r
# all()- 모두, 전부, any()- 어느거 하나라도 
x = 1:10 
x

x > 5 # 각각의 요소에 비교 연산을 수행 
```

```r
all(x>5) # 모든 것
any(x>5) # 어느 것 하나라도 
```

```r
name = c('홍길동','장보고','이순신','강감찬','김유신')
name2 = c('유관순', '장고보', '강감찬', '임꺽정', '홍길동')
age = c(17,21,19,31,22)
gender=factor(c('M','F','M','F','M'))
blood.type=factor(c('A','O','B','AB','O'))
```

```r
# 다른 이름으로 된 프레임을 병합할 때
pp1 = data.frame(name, age, gender)
pp1

pp2 = data.frame(name2, blood.type)
pp2
```

```r
# 공통인 부분만 merge
pp3 = merge(pp1, pp2, by.x='name', by.y='name2')
pp3
```

```r
# 다 merge
pp4 = merge(pp1, pp2, by.x='name', by.y='name2', all=TRUE)
pp4
```
