```r
# %in% : ~ 안에 포함여부를 판정하는 연산자

'a' %in% c('a','b','c') #안에 a가 있어서 T
'd' %in% c('a','b','c') #안에 d가 없어서 F
```

```r
#벡터를 집합을 취급, 집합연산자(합집합, 교집합, 차집합)
#집합(set)은 중복을 허락하지 않는다.

#합집합
union(c('a','b','c'), c('a','d')) #합하고 중복제거해서 출력

#교집합
intersect(c('a','b','c'), c('a','d')) #겹치는 값 출력

#차집합
setdiff(c('a','b','c'), c('a','d')) # 앞에 집합에서 교집합만 제거
```

```r
#집합의 비교연산 equals 의미
setequal(c('a','b','c'),c('a','d')) #원소값이 서로 같지 않음으로 F
setequal(c('a','b','c'),c('a','b','b','c')) #원소값이 서로 같으므로 T
```

```r
#all()- 모두, 전부 / any() -어느거 하나라도 
x=1:10
x
x>5
all(x>5) # 모든 x 값이 5보다 큽니까? F
any(x>5) # 어느 x 값중 5보다 큰 값이 하나라도 있습니까? T
```

```r
name = c('홍길동','장보고','이순신','강감찬','김유신')
name2 = c('유관순','장보고','강감찬','임꺽정','홍길동')
age = c( 17,21,19,31,22)
gender = factor(c('M','F','M','F','M'))
blood.type = factor(c('A','O','B','AB','O'))

pp1 = data.frame(name, age, gender)
pp1

pp2 = data.frame(name2, blood.type)
pp2

pp3 = merge(pp1, pp2, by.x = 'name' , by.y = 'name2') #by.x : 첫번째 데이터 키값 , by.y : 두번째 데이터 키값
pp3 #같은 값이 있는 것만 합쳐져서 출력

pp4 = merge(pp1, pp2, by.x = 'name' , by.y = 'name2', all= T)
pp4 #모두 합침. 대신 결측치가 생김
```

```r
#리스트의 넣을 값들 생성
a = c(1,2,3,4,5)
b = c(T,F,T,T,F,T)
c = data.frame(name = c('Alice','James','Merry'), age = c(23, 41, 19))

lis = list(a,b,c) #종류 상관 없이 묶을 수 있음


var_list = list(v1=a, v2=b, v3=c)# list에 변수명 지정
var_list

names(var_list) #객체의 변수명을 확인

names(var_list)[3] = 'df' #객체의 3번째 변수명 변경

names(var_list) = c('item1', 'item2', 'item3') #var_list의 변수명을 모두 변경
var_list
```

```r
#length() : 객체의 크기를 반환하는 함수
length(var_list) #리스트 카테고리가 3개가 있음
```

```r
#list의 str 과 summary
str(var_list) 
summary(var_list) #Class : 데이터를 담은 그릇 , Mode : 그릇에 담겨 있는 값의 형태
```

```r
var_list = list(v1=a, v2=b, v3=c)

#list의 카테고리 변경
var_list
var_list[['item3']] = c('A','B','C') # =를 써서 바꾸고 싶은 곳을 바꾸면 됨
var_list

var_list$item1 = NULL #삭제시
var_list

#새로운 카테고리 생성
var_list$new_item = data.frame(name=c('aaa','bbb'),age=c(12,29))
var_list
```

```r
#리스트와 각 요소에 동일한 함수를 반복문 없이 적용할때 사용하는 함수
#lapply(리스트, 적용할 함수) : 리스트로 결과 반환
#sapply(리스트, 적용할 함수) : 결과값이 하나면 벡터, 결과값이 같은 길이의 벡터면 행렬, 그외에는 리스트로 반환

korea_temp = list(경기 = c(-10, 2, 1, -2), 강원 = c(0, -4, -5, -10))
korea_temp

re_lap = lapply(korea_temp, mean) #korea_temp의 평균 온도를 리스트로 출력
re_lap
class(re_lap)

re_sap = sapply(korea_temp, mean) #korea_temp의 평균 온도를 벡터로 출력
re_sap
class(re_sap)

```
