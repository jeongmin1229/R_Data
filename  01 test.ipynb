```r
# 웹사이트에서 데이터셋을 가져와 작업을 수행 

tips = read.csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')
```

```r
str(tips)
```

```r
tail(tips, 20)
```

```r
# 기초 통계 분석 or 기술 통계 분석 
summary(tips)
```

```r
library(dplyr)
library(ggplot2)
```

```r
# 동석자수(size)의 분포 확인 - 빈도확인 
tips %>% ggplot(aes(size)) + geom_histogram()
```

```r
# total_bill과 tip과의 관계 - 상관관계 (산점도 or 산포도)
# tips 라는 데이터셋, ggplot함수로 변수 지정, point(산점도)로 변경 

tips %>% ggplot(aes(total_bill,tip)) + geom_point(aes(col=day, pch=sex), size = 3)
```
