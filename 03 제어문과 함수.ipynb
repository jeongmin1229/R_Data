### 제어문

```r
x = c(10, 20, 30, NA, 50)
x

# 벡터에 저장된 값이 40 미만인 값만 추출
x<40 # x의 원소의 값중 40 보다 작으면 T, 아니면 F / 결측값은 비교가 불가능하기 때문에 NA
x[x<40] #x의 원소 중 T인 값들과 겹치는 부분을 출력

x%%3 != 0 # %% : 나머지 구하는 함수 / 3의 배수가 아니면 T 아니면 F
x[x%%3 != 0] #위와 마찬가지로 T인 부분만 출력

is.na(x) #is.na(): 괄호 안에 있는 값이 결측값이면 T 아니면 F
x[is.na(x)] # T인 값만 출력
x[!is.na(x)] # F인 값만 출력 (반전 되서)
```

```r
# 논리연산자 : !(NOT) - 단항연산 / &(AND), |(OR)(아이 아닙니다) -이항연산

x%%2 == 0 & !is.na(x) # 2의 배수이면서, 결측값이 아닌 값을 T , 아니면 F

x[x%%2 == 0 & !is.na(x)] # T인 값을 출력
```

```r
#학점
score = readline ( "점수 입력 : ")
score = as.integer(score)

#as.character()
#as.numeric
#as


if(score>=90){
    print('A학점')  #첫번쨰 조건 만족시 출력
}else if(score>=80){
    print('B학점') #두번째 조건 만족시 출력
}else if(score>=70){
    print('C학점') #세번째 조건 만족시 출력
}else if(score>=60){
    print('D학점') #네번째 조건 만족시 출력
}else{
    print('F') #다섯번째 조건 만족시 출력
}
```

### 반복문

```r
# 1~10까지 수 출력
# repeat{반복처리할 구문}

num=1

repeat{ #무한 반복
    if(num > 10) { #10보다 커지면
        break #멈추고
    }else{ #그게 아니면
        print(num) #값을 프린트 한뒤
        num = num+1 #값에 +1을 합니다.
    } #이걸 다시 위로 올라가서 반복 하겠습니다.
}
```

```r
#for(개별 변수 in 집합 변수){ 반복처리할 구문}

for (i in 1:10){ # i는 1~10을 차례대로 반복할겁니다.
    print(i) # 현재 i 값을 출력 해줍니다.
} #다시 위로 가서 반복합니다.

```

```r
#while(조건식){참일때 반복할 구문}
i = 1

while(i<=10){ #i가 10 이하 일때 까지 아래 내용을 반복합니다.
    print(i) #현재 i 값을 출력 합니다.
    i=i+1 #i의 값을 +1 합니다.
} #다시 위로 가서 반복합니다
```

```r
# 1~100 사이의 짝수만 출력하는 구문

for(i in seq(100)){ 
    if(i%%2==0){
        print(i)
    }
}
```

```r
# 1부터 10 사이의 수중  소수만 출력하는 구문을 기술하시오.

for(i in 1:10){ #i를 1부터 10까지 실행
    count = 0 #카운트 포인트
    for(j in 1:i){ #j는 1부터 현재 i 값까지 반복
        if(i%%j == 0){ #i를 j로 나눴는데 나눠지면,
            count = count + 1 #카운트 값을 1
        } #count가 +1 된다는 것은 한번 나눠 졌다라는 말
    } #그렇다는 건 2번 나눠 졌는 것은 약수가 2개 라는 말 = 소수
    if(count == 2){ #위의 말과 동일할 때
        print(i) #i의 값을 출력해주세요.
    }
}#위에 반복
```

```r
#while문 구구단 출력
i=1
while(i<=10){
    print(paste(7, '*', i, '=', 7*i ))
    i = i+1
}
```

```r
j=1
while(j<10){
    cat(7, '*', j, '=', 7*j ,'\n')
    j = j+1
}
```

### 함수

```r
#점수가 80점 이상이면 합격, 그렇지 않으면 불합격

check <- function(score){  #check라는 함수를 다음과 같이 정의 (변수는 score)
        if(score>=80){ #score가 80이상이면
        print('합격') #합격 출력
    }else{
        print('불합격') #그 외엔 불합격 출력
    }
}

check(70)
check(88)
```

```r
# 입력받은 단수에따라 구구단을 출력하는 프로그램을 구현 하시오.
#단 입력은 사용자에 의해 받아서 처리할것

ggd<- function(){
    x = readline ( "출력할 구구단 : ")
    x = as.integer(x)
    j=1
    while(j<10){
        cat(x, '*', j, '=', x*j ,'\n')
        j = j+1
    }
}

ggd()
```
